```python
import numpy as np
import matplotlib.pyplot as plt
import h5py

## Stylistic only, if you don't have LaTeX locally installed, just disable
import scienceplots
plt.style.use('science')
```

# Benchmarking of the Q = 3^(M-1) Charge Sector

## N = 3

```python
N = 3
M = 6
J = 10

BT_data = np.load(f'BT_ED_data_N{N}_M{M}_J{J}.npz')
JL_data = h5py.File(f"ED_spectrum_N{N}_M{M}_J{J}.jld2", "r")

# 1. Sorted energies
energies = BT_data['energies']
energies_julia = JL_data['sorted_energies']
# 2. Access the entanglement entropy and corresponding eigenvalues
entropy = BT_data['entropy']

plt.figure(figsize = (6,6))
plt.scatter(np.arange(len(energies)), energies, s=20, label = 'BT')
print(len(energies))
plt.scatter(np.arange(len(energies_julia)), energies_julia, s=5, label = 'JL')
plt.ylabel(r'\textbf{Energy}')
plt.xlabel(r'\textbf{Eigenvalue\; Index}')
plt.title("Bo-Ting's (BT) ED Solver vs. Julia (JL) solver \nin $Q = 3^{M-1}$ Charge Sector" + f" (N={N}, M={M})")
# \n$Q = 3^{M-1}$
plt.legend()
plt.savefig(f'BT_JL_verification_N{N}_M{M}.pdf')
plt.show()

# plt.hist(model.eig_values_sorted[1:] - model.eig_values_sorted[:-1], 80)
# plt.xlim(0, 5)
# plt.ylim(0, 200)
# plt.show()

# plt.scatter(model.eig_values, model.entanglement_entropy, s=3)
# plt.axhline(y=np.log(2), linewidth=1, linestyle='dotted', color='purple')
# plt.show()
```
